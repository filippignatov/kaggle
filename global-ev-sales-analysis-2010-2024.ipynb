**Quickstart Exploratory Data Analysis (EDA)**

Welcome to the starter notebook for the Global Electric Vehicle Sales dataset. In this notebook, we will perform a quick visualization to identify which regions are leading the electric revolution in the final year of our dataset (2024)

# Load Libraries & Data
First, we import Pandas for data manipulation and Plotly for interactive visualizations

```python
import pandas as pd
import plotly.express as px

# Load the main sales dataset
# Path: /kaggle/input/[dataset-name]/ev_sales_main.csv
df = pd.read_csv('/kaggle/input/datasets/filippignatov/global-electric-vehicle-sales-2010-2024/ev_sales_main.csv')

# Display the first 5 rows to verify data integrity
df.head()
```

# Visualizing the 2024 Market Leaders

To understand the current state of the market, we filter for the year 2024 and visualize the top 10 regions by sales volume. We also color-code the bars by  **Vehicle_Type** (BEV vs. PHEV) to see the technology split.

```python
# Filtering for the most recent year in the dataset
top_2024 = df[df['Year'] == 2024].nlargest(10, 'Sales_Units')

# Creating an interactive bar chart
fig = px.bar(top_2024, 
             x='Region', 
             y='Sales_Units', 
             color='Vehicle_Type',
             title='🚀 Top 10 EV Markets in 2024',
             labels={'Sales_Units': 'Total Units Sold', 'Region': 'Market Region'},
             template='plotly_dark',
             barmode='stack')

fig.update_layout(xaxis={'categoryorder':'total descending'})

fig.show(renderer="iframe")
```

# Key Observations

Based on the visualization above, we can observe:


* Dominant Regions: Which markets have successfully scaled their EV adoption
* Tech Split: The ratio between pure Battery Electric Vehicles (BEV) and Plug-in Hybrids (PHEV)
* Market Maturity: The scale of sales in 2024 compared to the early 2010s baseline
